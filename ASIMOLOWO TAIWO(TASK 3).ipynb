```python
df= pd.read_csv("PizzaPlaceSales_merged.csv")
```

```python
#DATA CLEANING

#Converting of data
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["time"] = pd.to_datetime(df["time"], format="%H: %M: %S", errors="coerce").dt.time
df["price"] = pd.to_numeric(df["price"], errors= "coerce")
df["quantity"]= pd.to_numeric(df["quantity"], errors="coerce")

#Remove Duplicates
df = df.drop_duplicates()

#Standardizing text columns
df["name"] = df["name"].str.strip().str.lower()
df["category"] = df["category"].str.strip().str.lower()

#Creating Useful Colunms
df["sales"] = df["quantity"] * df["price"]
df["hour"] = pd.to_datetime(df["time"].astype(str), errors="coerce").dt.hour
df["day_of_week"] = df["date"].dt.day_name()
df["month"] = df["date"].dt.month_name()

#ANSWERING THE QUESTIONS
print("1. Total Revenue:", df["sales"].sum())
print("2. Total Quantity Sold:", df["quantity"].sum())
print("3. Total Orders:", df["order_id"].nunique())
print("4. Number of Pizza Types:", df["pizza_type_id"].nunique())
print("5. Average Pizza Price:", df["price"].mean())


peak_hours = (df.groupby("hour")["sales"].sum().sort_values(ascending=True))
print("6. Peak Hours pf Sales:", peak_hours.head())

print("7. Sales by Day of Week:")
print(df.groupby("day_of_week")["sales"].sum().sort_values(ascending=False))

print("8. Top 5 Bestselling Pizzas:")
print(df.groupby("name")["quantity"].sum().sort_values(ascending=False).head())

print("9. Monthly Sales Trend:")
print(df.groupby("month")["sales"].sum().sort_values(ascending=False))

print("10. Low Selling Pizza(Bottom 5):")
print(df.groupby("name")["quantity"].sum().sort_values().head(5))
```
